## Final Summary Tables and Export

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
def churn_summary(column):
    summary = df.groupby(column).agg(
        Total_Customers=('Churn', 'count'),
        Churned_Customers=('ChurnFlag', 'sum'),
        Churn_Rate=('ChurnFlag', 'mean')
    ).reset_index()

    summary['Churn_Rate'] = summary['Churn_Rate'] * 100
    return summary.sort_values(by='Churn_Rate', ascending=False)
```

```python
# Recreate important helper columns before summary tables

# 1. Create ChurnFlag
df['ChurnFlag'] = df['Churn'].map({'Yes': 1, 'No': 0})

# 2. Create TenureGroup again, just in case
df['TenureGroup'] = pd.cut(
    df['tenure'],
    bins=[-1, 12, 24, 48, 72],
    labels=['0-12 months', '13-24 months', '25-48 months', '49-72 months']
)

# 3. Create MonthlyChargeGroup again, just in case
df['MonthlyChargeGroup'] = pd.cut(
    df['MonthlyCharges'],
    bins=[0, 35, 70, 120],
    labels=['Low charge', 'Medium charge', 'High charge']
)

# 4. Check that ChurnFlag was created correctly
df[['Churn', 'ChurnFlag', 'TenureGroup', 'MonthlyChargeGroup']].head()
```
