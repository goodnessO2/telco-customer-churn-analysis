## Final Summary Tables and Export

```python
df['Churn'].unique()
```

```python
if df['Churn'].dtype == 'object':
    df['ChurnFlag'] = df['Churn'].str.strip().map({'Yes': 1, 'No': 0})
else:
    df['ChurnFlag'] = df['Churn'].astype(int)

# Confirm ChurnFlag is correct
df[['Churn', 'ChurnFlag']].head()
```

```python
df['ChurnFlag'].isnull().sum()
```

```python
df['ChurnFlag'].value_counts()
```

```python
# Create tenure groups
df['TenureGroup'] = pd.cut(
    df['tenure'],
    bins=[-1, 12, 24, 48, 72],
    labels=['0-12 months', '13-24 months', '25-48 months', '49-72 months']
)

# Create monthly charge groups
df['MonthlyChargeGroup'] = pd.cut(
    df['MonthlyCharges'],
    bins=[0, 35, 70, 120],
    labels=['Low charge', 'Medium charge', 'High charge']
)

df[['tenure', 'TenureGroup', 'MonthlyCharges', 'MonthlyChargeGroup']].head()
```

```python
def churn_summary(column):
    summary = df.groupby(column, observed=False).agg(
        Total_Customers=('ChurnFlag', 'count'),
        Churned_Customers=('ChurnFlag', 'sum')
    ).reset_index()

    summary['Churn_Rate'] = (
        summary['Churned_Customers'] / summary['Total_Customers']
    ) * 100

    summary['Churn_Rate'] = summary['Churn_Rate'].round(2)

    return summary.sort_values(by='Churn_Rate', ascending=False)
```

```python
overall_churn_summary = df['ChurnFlag'].value_counts().reset_index()
overall_churn_summary.columns = ['ChurnFlag', 'Customer_Count']

overall_churn_summary['Churn_Status'] = overall_churn_summary['ChurnFlag'].map({
    0: 'No Churn',
    1: 'Churn'
})

overall_churn_summary['Percentage'] = (
    overall_churn_summary['Customer_Count'] / len(df)
) * 100

overall_churn_summary['Percentage'] = overall_churn_summary['Percentage'].round(2)

overall_churn_summary = overall_churn_summary[
    ['Churn_Status', 'Customer_Count', 'Percentage']
]
```

```python
contract_summary = churn_summary('Contract')
tenure_summary = churn_summary('TenureGroup')
monthly_charge_summary = churn_summary('MonthlyChargeGroup')
internet_service_summary = churn_summary('InternetService')
payment_method_summary = churn_summary('PaymentMethod')
```
